```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams['figure.dpi'] = 120
```

#### mock data

```python
x = np.arange(0,10, 0.25)
y = np.sin(x) 
y2 = np.cos(x)
```

## Matlab-like Quick plots

```python
plt.plot(x,y);
```

```python
plt.plot(x,y, ':g', label='sin');
plt.plot(x,y2, '^b',label='cos');

## above is equal to:
## plt.plot(x, y, ':g', x,y2, '^b');

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()
plt.show()
```

## "Subplots"

```python
fig, axes = plt.subplots(ncols=2, nrows=2,figsize=(8, 6))
ax = axes[0][0]
ax.plot(x,y);

ax = axes[0][1]
ax.hist(y);
fig.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```
